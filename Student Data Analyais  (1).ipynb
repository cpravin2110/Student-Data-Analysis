```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop the Unnamed col
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
print(df.head())
```

```python
plt.figure(figsize=(3,5))
pp=sns.countplot(data=df,x="Gender", palette="cool_r")
pp.bar_label(pp.containers[0])
pp.bar_label(pp.containers[1])
plt.title("Gender Disribution")
plt.show()
```

```python
# from parents education background we check the impact on students scores.

GB=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(GB)
```

```python
sns.heatmap(GB,annot=True)
plt.show()
```

```python
# from parents Relation Status we check the impact on students scores.

GB1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(GB1)
```

```python
sns.heatmap(GB1,annot=True)
plt.show()
```

```python
# Number of students present in perticular group

print(df["EthnicGroup"].unique())
```

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
 
l=["group A","group B","group C","group D","group E"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.title('Distibution of Groups')
plt.show()
```

```python
# Number of students play sports by Gender and Sport Categary

ppp=sns.countplot(data=df,x="PracticeSport",hue="Gender",palette="bright")
ppp.bar_label(ppp.containers[0])
ppp.bar_label(ppp.containers[1])
plt.title("students play sports by Gender and Sport Categary")

plt.show()
```

```python
# Student Transport mode by Gender

ppp=sns.countplot(data=df,x="TransportMeans",hue="Gender",palette="cool_r")
ppp.bar_label(ppp.containers[0])
ppp.bar_label(ppp.containers[1])
plt.title("Student Transport mode by Gender")

plt.show()
```
